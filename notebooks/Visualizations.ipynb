# Visualizations

This shows some basic visualizations offered by iquaflow.

They use dataframes as inputs, these dataframes can be generated by the **get_df** method from the **ExperimentInfo**

```python
from examples import *
```

```python
df1 = sample_df("agg1")
ev = ExperimentVisual(df1)
```

```python
ev.visualize(
    plot_kind='lineplot',
    xvar='ds_modifier',
    yvar='val_rmse',
    legend_var='min_size',
    title='Line Plot Example'
)
```

```python
ev.visualize(
    plot_kind='bars',
    xvar='ds_modifier',
    yvar='val_rmse',
    legend_var='min_size',
    title='Bars Plot Example'
)
```

```python
ev.visualize(
    plot_kind='scatter',
    xvar='ds_modifier',
    yvar='val_rmse',
    legend_var='min_size',
    title='Scatter Plot Example'
)
```

```python
df2 = sample_df("agg2")
ev2 = ExperimentVisual(df2)
```

```python
ev2.visualize(
    plot_kind='lineplot',
    xvar='ds_modifier',
    yvar='val_rmse',
    legend_var='min_size',
    title='Line Plot Example',
    plot_mean_std=True,
)
```

```python
df3 = sample_df("ROC")
ev3 = ExperimentVisual(df3)
```

```python
ev3.visualize(
    xvar="Precision",
    yvar="Recall",
    legend_var="ds_modifier",
    title="ROC"
)
```
